## Preview of Data science tools
To learn further, here are some cool packages to know about:

Numpy: Matrices  
Pandas: Labeled columns and visualization <br/>
Matplotlib: better plotting  
SciPy: Scientific Python 


```python
#$ conda install numpy scipy pandas matplotlib scikit-learn

import numpy as np
# Numerical Python
# Numpy allows matrix manipulation
x = np.arange(1,10)
x
#NumPy's arrays offer efficient storage of data and element-wise operations
x**2
# for Python you'd have to use
[val ** 2 for val in range (1,10)]
#unlike Python lists limited to 1D, these are M-D
M = x.reshape((3,3))
M.T #transpose
np.dot(M, [5,6,7]) #matrix-vector products
np.linalg.eigvals(M)


import pandas as pd
# Labeled Column-oriented data
# built on top of Numpy
df = pd.DataFrame({'label': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'value': [1, 2, 3, 4, 5, 6]})
df['label'].str.lower()
df['value'].sum()
df.groupby('label').sum()



import matplotlib.pyplot as plt
# Matlab-stype scientific visualisation
#may have to run the following in the Jupyter notebook before hand: %matplotlib notebook
plt.style.use('ggplot') #make graphis in the style of R's ggplot
x = np.linspace(0,10) #range of values from 0 to 10 using Numpy
y = np.sin(x)
plt.plot(x,y)
# there is a way to make the chart interactive...?



from scipy import interpolate
# scientific functionalities leveraging Numpy:
#     scipy.fftpack: Fast Fourier transforms
#     scipy.integrate: Numerical integration
#     scipy.interpolate: Numerical interpolation
#     scipy.linalg: Linear algebra routines
#     scipy.optimize: Numerical optimization of functions
#     scipy.sparse: Sparse matrix storage and linear algebra
#     scipy.stats: Statistical analysis routines
func = interpolate.interp1d(x, y, kind='cubic')
x_interp = np.linspace(0, 10, 1000)
y_interp = func(x_interp)
plt.figure() #new figure
plt.plot(x,y,'o')
plt.plot(x_interp, y_interp)
```
